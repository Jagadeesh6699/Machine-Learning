```python
plt.figure(figsize=[8,3])
x=[1,2,3,4,5]
y=[10,15,20,10,40]
y1=np.tan(x)

plt.plot(x,y,color='b',marker="^",markersize=14,linestyle="--",label="temp")
plt.plot(x,y1,color='green',marker='o',label="tan")
# Line plot
plt.bar(x,y,color='pink')
plt.xlabel("Hours")           # getting x axis label
plt.ylabel("temp")            # diplay y axis variable
plt.title("Hourly temparature chart")
plt.legend()
plt.xlim((0,7))   #manipulate the range of x axis
plt.ylim((-5,50))
plt.xticks(x,labels=["Excel","SQL","Python","Machine Learning","DL"],rotation=90)

for i,j in zip(x,y):
    plt.text(x=i-0.1,y=j+2.5,s=str(j))
    #plt.text(x=i,y=j+5,s=(str(i)+","+str(j)))

plt.grid()
plt.show()
plt.savefig("export_barchart.png")
```

```python
#pie charts are useful to visualize the % distribution of a single
#categorical variable
plt.figure(figsize=(5,3))
import matplotlib.pyplot as plt

#slice will be plo,lables tted counter-clockwise:
labels=['SQL','Python','R','Tableau']
sizes=[75,156,40,30]
#sizes = [14.5,45.5,30,10]
explode = [0,0,0.2,0] # to seperate with a gap
plt.pie(sizes,labels=labels,explode=explode,autopct="%1.1f%%",
       colors=["r","g","purple","y"],
       #autopct=to specify the percentage format
       shadow=True,startangle=90)
plt.show()
```

# Probability Distribution

## Bernoullis Distribution

```python
from scipy import stats
```

```python
x=[1,0]
p=0.7
q=1-p
stats.bernoulli.pmf(x,p)
```

## Binomial Distribution

```python
prob=stats.binom.pmf(10,30,0.2)
print(f"required prob is :{prob*100}")
```

```python
prob=stats.binom.cdf(1,30,0.2)
prob*100
```

```python
req_prob=1-stats.binom.cdf(10,30,0.2)
req_prob*100
```

## Poisson Distribution

```python
prob=stats.poisson.pmf(30,25) #exact 30 return p(x=30)
prob
```

```python
prob=1-stats.poisson.cdf(30,25)   #greater than 30 returns
prob
```
